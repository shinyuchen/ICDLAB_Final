```python
import random
gen_num = 20

input_list = []
output_list = []
for i in range(gen_num):
  temp_in = random.randint(-64, 63)
  if(temp_in >= 0):
    output_list.append(temp_in)
  else:
    output_list.append(0)
  input_list.append(temp_in)
```

```python
def int_to_twos_complement(n, num_bits):
    """
    Converts integer n to its two's complement binary representation using num_bits bits.
    """
    # Check if n is negative
    negative = n < 0

    # Compute the absolute value of n and convert it to binary
    abs_n = abs(n)
    binary = bin(abs_n)[2:]

    # Pad the binary string with zeros if necessary to make it num_bits long
    if len(binary) < num_bits:
        binary = '0' * (num_bits - len(binary)) + binary

    # Invert the bits if n is negative
    if negative:
        inverted = ''
        for bit in binary:
            if bit == '0':
                inverted += '1'
            else:
                inverted += '0'
        binary = inverted

    return binary
```

```python
bit_len = 8

with open('./input.txt', 'w') as f:
  for each in input_list:
    each = int_to_twos_complement(each, bit_len)
    print(each)
    f.write(str(each))
print("====================================")
with open('./output.txt', 'w') as f:
  for each in output_list:
    each = int_to_twos_complement(each, bit_len)
    print(each)
    f.write(str(each))
```
