```python
import torch
```

```python
class one_layer_net(torch.nn.Module):    
    # Constructor
    def __init__(self, input_size, weight=None, bias=None):
        super(one_layer_net, self).__init__() 
        self.batch_norm_1 = torch.nn.BatchNorm2d(input_size)
        if weight != None:
          self.batch_norm_1.weight = torch.nn.parameter.Parameter(weight)
        if bias != None:
          self.batch_norm_1.bias = torch.nn.parameter.Parameter(bias)
    def forward(self, x):
        y_pred = self.batch_norm_1(x)
        return y_pred
```

```python
batch_size = 1
feature_num = 1
height = 1
width = 5

new_weight = torch.rand(batch_size, feature_num)
new_bias = torch.rand(batch_size, feature_num)

model = one_layer_net(feature_num, weight = new_weight, bias = new_bias)
# model = one_layer_net(feature_num)
print('Network Structure : torch.nn.Linear(2,1) :\n',model.batch_norm_1)
print('Weight Of Network :\n',model.batch_norm_1.weight)
print('Bias Of Network :\n',model.batch_norm_1.bias)
```

```python

input = torch.randn(batch_size, feature_num, height, width)
print(input)
out = model(input)
print("Model Output")
print(out)

firstDimenMean=torch.Tensor.mean(input[0][0])
firstDimenVar=torch.Tensor.var(input[0][0], False)
print(model.batch_norm_1.eps)
print(firstDimenMean)
print(firstDimenVar)
print(torch.pow(firstDimenVar+model.batch_norm_1.eps, 0.5))
```

```python
batchnormone=((input[0][0][0][0]-firstDimenMean) /(torch.pow(firstDimenVar+model.batch_norm_1.eps, 0.5) )) * model.batch_norm_1.weight[0] + model.batch_norm_1.bias[0]
print(batchnormone)
```

```python
weight_to_write = model.batch_norm_1.weight.tolist()[0][0]
bias_to_write = model.batch_norm_1.bias.tolist()[0][0]
input_to_write = input.tolist()[0][0]
golden_to_write = out.tolist()[0][0]
```
